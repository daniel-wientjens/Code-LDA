```python
# the NMF decomposes this Term Document matrix into the product of 2 smaller matrices: W and H
n_dimensions = 10 # This can also be interpreted as topics in this case. This is the "beauty" of NMF. 10 is arbitrary
nmf = NMF(n_components=n_dimensions, random_state=42, alpha=.1, l1_ratio=.5).fit(datan) 
# fixing the parameter "random_state=42" will result in having the same result if you re-run the algorithm again. 42 is arbitrary
```

```python
W = nmf.fit_transform(datan)
H = nmf.components_
```

```python
from gensim import models, corpora
from gensim.utils import simple_preprocess
from gensim.models import Phrases
import stop_words
import tqdm
```

```python
STOPWORDS = stop_words.get_stop_words(language='en')
```

```python
def tokenize(text):
    return [token for token in simple_preprocess(text) if token not in STOPWORDS]
```

```python
datan2 = pd.read_csv('/Users/Daniel/Documents/Courses X/Capgemini/Data/datanstuff.csv')
```

```python
datan2
```

```python
list_tokens_all_articles = [tokenize(comments) for comments in datan2.text.tolist()]
```

```python
# Use bigram and trigram to catch combination of 2/3 words that have a specific meaning together
bigram = Phrases(list_tokens_all_articles, min_count=5)
trigram = Phrases(bigram[list_tokens_all_articles], min_count=3)
```

```python
tokens = list(trigram[bigram[list_tokens_all_articles]])
```

```python
id2word_newspaper = corpora.Dictionary(tokens)
print(id2word_newspaper)
```

```python
# ignore words that appear in less than 5 documents or more than 30% documents
id2word_newspaper.filter_extremes(no_below=5, no_above=0.3)
print(id2word_newspaper)
```

```python
corpus = [id2word_newspaper.doc2bow(tok) for tok in tokens]
```

```python
# Run the LDA (computation time should be between 5 to 60 seconds)

# choose the number of topics => to find a "good" number of topics, try multiple values and see which one is the best
# optionally: input alpha and eta to influence how topics are distributed across documents, 
#  and how words are distributed across topics
#  the syntax is the following
#  alpha is a vector of size the number of documents, and eta's size is the number of words
#  alpha = [0.01] * id2word_newspaper.num_docs for instance
#  eta = [0.01] * len(id2word_newspaper.keys())

num_topics = 40

# Below without alpha nor eta
%time lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=id2word_newspaper, passes=4)

# Below with alpha and eta
# %time lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=id2word_newspaper, passes=4, 
# 
```

```python
num_documents = 5 # increase this number to check how topics are distributed across more than the first 5 documents
for i in range(5):
    print(lda_model[corpus[i]]) # what proportion of topics in each document
```

```python
lda_model.show_topics(num_topics=num_topics, num_words=10, formatted=True)
```

```python
import math # for rounding
[(topic[0], math.ceil(topic[1]*100)/100) for topic in lda_model[corpus[i]]]
```

```python
# For each article, print the beginning of the article, the topic allocation for this document, 
# and the 5 main keywords of the first 2 topics of this document

topics = lda_model.show_topics(num_topics=num_topics, num_words=5, formatted=True)

for i,article in enumerate(datan2.text.tolist()):
    topics_in_article = lda_model[corpus[i]]
    topics_in_article = [(topic[0], math.ceil(topic[1]*100)/100) for topic in topics_in_article] # rounded, for better printing
    key_words_topic = []
    
    for topic in range(min(len(topics_in_article), 2)):
        id_topic = topics_in_article[topic][0] # id of the topic
        key_words_topic.append(topics[id_topic][1])
    
    print("article #"+str(i)+": "+article[:100].replace('\n','') + '\n' + str(topics_in_article) + '\n' +
          '\n'.join(["keywords topic: "+kw for kw in key_words_topic]))
    print()
```
